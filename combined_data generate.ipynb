```python
import pandas as pd

# 파일 경로 정의
accident_data_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/accident_data_with_engineered_features.csv'
cctv_data_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/daegu_cctv_cleaned.csv'
child_protection_area_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/daegu_child_protection_area_cleaned.csv'
parking_lot_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/daegu_parking_lot_cleaned.csv'
security_light_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/daegu_security_light_cleaned.csv'

# 각 데이터셋을 불러오기
accident_data = pd.read_csv(accident_data_path)
cctv_data = pd.read_csv(cctv_data_path)
child_protection_area = pd.read_csv(child_protection_area_path)
parking_lot = pd.read_csv(parking_lot_path)
security_light = pd.read_csv(security_light_path)

# 데이터셋의 첫 5줄을 출력하여 구조 파악
accident_data_head = accident_data.head()
cctv_data_head = cctv_data.head()
child_protection_area_head = child_protection_area.head()
parking_lot_head = parking_lot.head()
security_light_head = security_light.head()

(accident_data_head, cctv_data_head, child_protection_area_head, parking_lot_head, security_light_head)
```

```python
# 각 데이터셋의 필요한 특징(Feature) 추출 및 병렬 데이터 구조 생성

# 1. 사고 데이터에서 필요한 특징 추출
accident_features = accident_data[['사고일시', '시군구', '사고유형', '사망자수', '중상자수', '경상자수']]

# 2. CCTV 데이터에서 필요한 특징 추출
cctv_features = cctv_data[['시군구명', '위도', '경도']]

# 3. 아동 보호구역 데이터에서 필요한 특징 추출
child_protection_features = child_protection_area[['시설종류', '위도', '경도']]

# 4. 주차장 데이터에서 필요한 특징 추출
parking_lot_features = parking_lot[['주차장명', '위도', '경도']]

# 5. 보안등 데이터에서 필요한 특징 추출
security_light_features = security_light[['보안등위치명', '위도', '경도']]

# 각 데이터셋의 특징을 합친 새로운 데이터 프레임 생성
combined_features = pd.concat([accident_features, cctv_features, child_protection_features, parking_lot_features, security_light_features], axis=1)

# 생성된 병렬 데이터 구조 확인
combined_features.head()
```

```python
# 각 특징의 데이터 타입과 결측치 상태 확인

# 데이터 타입 확인
data_types = combined_features.dtypes

# 결측치 확인
missing_values = combined_features.isnull().sum()

(data_types, missing_values)
```

```python
# 추가적인 처리 진행: 결측치 처리, 데이터 타입 변환, 특징 스케일링

# 1. 데이터 타입 변환: '사고일시'를 datetime 타입으로 변환
combined_features['사고일시'] = pd.to_datetime(combined_features['사고일시'], errors='coerce')

```

```python
# 범주형 데이터의 결측치는 '미확인' 또는 가장 빈번한 값으로 대체
for column in combined_features.select_dtypes(include=['object']).columns:
    combined_features[column].fillna(combined_features[column].mode()[0], inplace=True)

```

```python
# 3. 특징 스케일링: 수치형 데이터에 대해 표준화 진행
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_columns = combined_features.select_dtypes(include=['float64', 'int64']).columns
combined_features[numeric_columns] = scaler.fit_transform(combined_features[numeric_columns])

# 처리 후의 데이터 확인
combined_features.head()
```

```python
# CSV 파일로 저장
combined_features.to_csv('C:/Users/82106/Desktop/데이터분석 프로젝트/대구 교통사고 피해 예측 AI 경진대회/open/open/combined_data.csv', index=False)
```
